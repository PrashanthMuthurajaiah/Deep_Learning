1. cv2.line() : Used to draw line on an image.
2. cv2.rectangle() : Used to draw rectangle on an image.
3. cv2.circle() : Used to draw circle on an image.
4. cv2.putText() : Used to write text on image.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = np.zeros((400, 400, 3), dtype = "uint8")
plt.imshow(img)
plt.title("Black Image")
plt.show()
```

# CV2 Line

cv2.line(imageObjectName, (‘start_coordinates’), (‘end_coordinates’), (‘color_in_bgr’), ‘line_thickness’)

```python
line_image = cv2.line(img, (20, 160), (100, 160), (0, 0, 255), 10)
plt.imshow(line_image)
plt.title("Line Image")
plt.show()
```

cv2.rectangle(imageObjectName, (‘top_left_vertex_coordinates’), (‘lower_right_vertex_coordinates’), (‘stroke_color_in_bgr’), ‘stroke_thickness’)

```python
rectangle_image = cv2.rectangle(img, (30,30),(300, 200), (0,0,225), 10)
```

```python
plt.imshow(rectangle_image)
plt.title("Rectangle Image")
plt.show()
```

Draw a circle

cv2.circle(imageObjectName, (‘center_coordinates’), (‘circle_radius’), (‘color_in_bgr’), ‘stroke_thickness’)

```python
circle_image = cv2.circle(img, (200, 200), 80, (255, 0, 0), 3) 
plt.imshow(circle_image)
plt.title("circle_image")
plt.show()
```

Put a text on the Image

cv2.putText(imageObjectName, ‘TextContent’, (‘text_starting_point_coordinates’), ‘fontToBeUsed’, ‘font_size’, (‘text_color’, ‘text_thickness’, ‘line_type’)

```python
font = cv2.FONT_HERSHEY_SIMPLEX 
text_mg = cv2.putText(img, 'Sample', (50, 50), 
            font, 0.8, (0, 255, 0), 2, cv2.LINE_AA) 
plt.imshow(text_mg)
plt.title("text_mg")
plt.show()
```

# Image Translation

Translation refers to the rectilinear shift of an object i.e. an image from one location to another.If we know the amount of shift in horizontal and the vertical direction, say (tx, ty).

where tx denotes the shift along the x-axis and ty denotes shift along the y-axis i.e. the number of pixels by which we need to shift about in that direction.
Now, we can use the cv2.wrapAffine() function to implement these translations.



```python
image = cv2.imread('./Cat.jpg') 
```

```python
# Store and height and width of image
height, width = image.shape[:2]
```

```python
# Taking the quarter of height and weight 
quarter_height, quarter_width = height / 4, width / 4
```

```python
T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]]) 
```

```python
# We use warpAffine to transform 
# the image using the matrix, T 
img_translation = cv2.warpAffine(image, T, (width, height)) 
```

```python
plt.imshow(img_translation)
plt.title("Image translation")
plt.show()
```
