To read the contents of an image, we have a function cv2.imread()

1. cv2.IMREAD_COLOR : Loads image in color. It is used by default.
2. cv2.IMREAD_GRAYSCALE : Loads image in gray mode.
3. cv2.IMREAD_UNCHANGED : Loads image as it is, including

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
# Load the image in color mode.

img_color = cv2.imread('./Cat.jpg', cv2.IMREAD_COLOR)

plt.imshow(img_color)
plt.title("Colored Cat")
plt.show()
```

```python
# Load the image in grayscale mode.

img_gray = cv2.imread('./Cat.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(img_gray)
plt.title("Grayscaled Cat")
plt.show()
```

```python
# Load the image in unchanged mode.

img_unchanged = cv2.imread('./Cat.jpg', cv2.IMREAD_UNCHANGED)

plt.imshow(img_unchanged)
plt.title("Unchanged Cat")
plt.show()
```

```python
# Saving the grayscaled cat

cv2.imwrite('Gray_cat.jpg', img_gray)
```
